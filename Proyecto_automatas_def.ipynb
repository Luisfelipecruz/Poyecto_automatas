```
GR = {'b': ['g', 'r', '6', 'f'], 'f': ['b', 'p'],'p': ['f', 'q', 'a'],
        'q': ['p', 'y'],'y': ['q', 'x','w'],'w': ['y', 'k'],'k': ['w', 'e'], 
        'e': ['k', '5'],'5': ['e', 'j'], 'j': ['5', 'm'],'m': ['j','n'],'n': ['m', 'i','4', 'h'],
        'i': ['n', 'a'],'a': ['i', 'p'],'4': ['n'],'h': ['n', 'c'], 'c': ['h', 'o'],'o': ['c', 'd'],'d': ['o', 's'],
        's': ['d', 'u'],'u': ['s','3'],'3': ['u', 'v'], 'v': ['3', 'z'], 'z': ['v', 't'],'t': ['z', '2'],
        '2': ['t', 'l', '7'],'l': ['2', 'g'],'g': ['l', 'b'],'7': ['2', 'r'],'r': ['7', 'b'],'6': ['b','x'], 'x': ['6', 'y']
     }  
```

```
def find_all_paths(graph, start, end, path=[],):
        path = path + [start]
        if start == end:
            return [path]
        if not graph.__contains__(start):
            return []
        paths = []
        for node in graph[start]:
            if node not in path:
                newpaths = find_all_paths(graph, node, end, path)
                for newpath in newpaths:
                    paths.append(newpath)                    
        return paths   
```
