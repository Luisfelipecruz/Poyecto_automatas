# Procesar los datos del censo de 2018 para el departamento de Santander y unirlos con el shapefile

```
import pandas as pd 
censo_2018_santander = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/proyecto_automatas/municipios_shapefile/CNPV2018_5PER_A2_68.CSV") 
censo_2018_santander =censo_2018_santander.rename(columns={"U_MPIO": "MPIO_CCDGO"})
censo_2018_santander['MPIO_CCDGO']=censo_2018_santander['MPIO_CCDGO'].astype(int)
print('numero de filas: ' + str(len(censo_2018_santander)))
censo_2018_santander.head()
```

```
Santander_shape_censo = shapefile_santander.merge(censo_2018_santander, on='MPIO_CCDGO')
print('numero de filas: ' + str(len(Santander_shape_censo)))
Santander_shape_censo.head()
```

Shape de los municipios con todos los datos del censo del 2018

```
poblacion_x_municipio = Santander_shape_censo.groupby(['Nombre municipio']).size().reset_index(name='Poblacion_2018')
pbc_x_muni_orden = poblacion_x_municipio.sort_values(by='Poblacion_2018', ascending=False)
print('numero de filas: ' + str(len(pbc_x_muni_orden)))
pbc_x_muni_orden.head()
```

```
Snt_mun_shp_pobla = shapefile_santander.merge(poblacion_x_municipio, on='Nombre municipio')
print('numero de filas: ' + str(len(Snt_mun_shp_pobla)))
Snt_mun_shp_pobla.head()
```

```
lista_mun_sant = []
for row in Snt_mun_shp_pobla.iterrows():
  lista_mun_sant.append(row[1]["Nombre municipio"])
print(lista_mun_sant)
```

```
import matplotlib.pyplot as plt
import matplotlib.colors as colors
norm = colors.Normalize(vmin=Snt_mun_shp_pobla["Poblacion_2018"].min(), vmax=Snt_mun_shp_pobla["Poblacion_2018"].max())
cbar = plt.cm.ScalarMappable(norm=norm, cmap='RdBu')
fig, ax = plt.subplots(figsize=(15, 7))
Snt_mun_shp_pobla.plot(column='Poblacion_2018', cmap='RdBu', legend=False, ax=ax)
ax_cbar = fig.colorbar(cbar, ax=ax)
ax_cbar.set_label('Distribucion poblacion del departamento de Santander')
```

# Acceder a los datos de covd actualizados y unirlos con el shapefile


```
import requests
url = 'https://www.datos.gov.co/api/views/gt2j-8ykr/rows.csv?accessType=DOWNLOAD'
r = requests.get(url, allow_redirects=True)
open('Casos_positivos_de_COVID-19_en_Colombia.csv', 'wb').write(r.content)
data = pd.read_csv('Casos_positivos_de_COVID-19_en_Colombia.csv', sep=',')
data['Código ISO del país'] = data['Código ISO del país'].fillna(169)
data['Pertenencia étnica'] = data['Pertenencia étnica'].fillna(0)
data['Nombre del país'] = data['Nombre del país'].fillna('Colombia')
data['Fecha de recuperación'] = data['Fecha de recuperación'].fillna('No recuperado')
data['Tipo de recuperación'] = data['Tipo de recuperación'].fillna('No recuperado')
data['Fecha de inicio de síntomas'] = data['Fecha de inicio de síntomas'].fillna('No hay registro')
data['Fecha de muerte'] = data['Fecha de muerte'].fillna('No ha muerto')
data = data.drop(columns='Nombre del grupo étnico')
data = data.loc[data["Nombre departamento"] == "SANTANDER"]
print('numero de filas: ' + str(len(data)))
data.head()
```

```
lista_mun_activos=[]
data_covid_Santander = data.groupby(['Nombre municipio','Recuperado']).size().reset_index(name='Casos')
for row in data_covid_Santander.iterrows():
  if (row[1]["Recuperado"] == 'Activo'):
    lista_mun_activos.append(row[1]["Nombre municipio"])
print(len(lista_mun_sant)-len(lista_mun_activos))
print((lista_mun_sant))
print((lista_mun_activos))
```

```
lista_faltantes  = []
for item in lista_mun_sant:
  if item not in lista_mun_activos:
    lista_faltantes.append(item)
print('Los siguientes municipios no presentan casos activos de covid el dia de hoy, en total son : '+ str(len(lista_faltantes)))
for i in range(len(lista_faltantes)):
  print((i+1, lista_faltantes[i]),end = '')
  if i% 5 == 0 and i != 0:
    print('')
```

```
data_covid_Santander = data.groupby(['Nombre municipio','Recuperado']).size().reset_index(name='Casos')

for i, row in data_covid_Santander.iterrows():
    if row["Recuperado"] != 'Activo':
      data_covid_Santander.at[i,"Recuperado"] = 'NaN'

data_covid_nan=data_covid_Santander[data_covid_Santander.Recuperado == 'NaN']
data_covid_nan = data_covid_nan.groupby(['Nombre municipio','Recuperado']).size().reset_index(name='Casos')
data_covid_activos=data_covid_Santander[data_covid_Santander.Recuperado == 'Activo']

for i, row1 in data_covid_nan.iterrows():
  for j, row2 in data_covid_activos.iterrows():
    if (row1["Nombre municipio"] == row2["Nombre municipio"] and row1["Recuperado"] != row2["Recuperado"]):
        data_covid_nan.at[i,"Recuperado"] = row2["Recuperado"]
        data_covid_nan.at[i,"Casos"] = row2["Casos"]
    else:
      continue
for i, row1 in data_covid_nan.iterrows():
  if (row1["Recuperado"] == 'NaN'):
      data_covid_nan.at[i,"Casos"] = int(0)
data_covid_Santander=data_covid_nan.sort_values(by='Casos', ascending=False)
print('numero de filas: ' + str(len(data_covid_Santander)))
data_covid_Santander.head()
```

```
Santander_casos_activos = Snt_mun_shp_pobla.merge(data_covid_Santander, on='Nombre municipio')
print('numero de filas: ' + str(len(Santander_casos_activos)))
Santander_casos_activos=Santander_casos_activos.sort_values(by='Casos', ascending=False)
Santander_casos_activos.head()
```

```
norm = colors.Normalize(vmin=Santander_casos_activos["Casos"].min(), vmax=Santander_casos_activos["Casos"].max())
color_map = plt.cm.get_cmap('RdYlGn')
fig, ax = plt.subplots(figsize=(15, 7))
reversed_color_map = color_map.reversed()
Santander_casos_activos.plot(column='Casos', cmap=reversed_color_map, legend=False, ax=ax)
cbar = plt.cm.ScalarMappable(norm=norm, cmap=reversed_color_map)
ax_cbar = fig.colorbar(cbar, ax=ax)
ax_cbar.set_label('Casos activos en departamento de Santander')
```
